## 作業目標：搭建一個bag of words模型

---

```python
import pandas as pd
import nltk
#nltk.download()
import numpy as np
dataset = pd.read_csv('./datasets/Restaurant_Reviews.tsv', delimiter='\t', quoting=3)
corpus = dataset['Review'].values
```

```python
### 從文本中取出所有單字
```

```python
whole_words = []
for sentence in corpus:
    tokenized_sentence = nltk.word_tokenize(sentence)
    for word in tokenized_sentence:
        whole_words.append(word)
```

```python
whole_words[:10]
```

### 移除重複單字

```python
whole_words = list(set(whole_words))
print('共有{}個單字'.format(len(whole_words)))
```

### 建立字典使每一個單字有對應數值

```python
word_index = {}
index_word = {}
n = 0

for word in whole_words:
    # code
    word_index[word] = n
    index_word[n] = word
    n += 1
```

## 轉換句子為bag of words型式

```python
def _get_bag_of_words_vector(sentence, word_index_dic, whole_words):
    sentence = sentence
    '''創建一個vector'''
    vector = [0] * len(whole_words)
    for word in nltk.word_tokenize(sentence):
        if word in whole_words:
            # code
            vector[word_index_dic[word]] += 1
    return vector
    
```

```python
_get_bag_of_words_vector('Wow... Loved this place.', word_index, whole_words)
```
